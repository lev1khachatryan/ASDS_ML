```python
import glob
def read_file_to_dict(path = r'Data',file_type = 'csv' ):
    Files = glob.glob(path + "/*." + file_type)
    dict_of_files = {}
    if file_type == 'csv':
        for file in Files:
            df = pd.read_csv(file, index_col=False)
            dict_of_files[str(file)[5:-4]] = df
    elif file_type == 'xlsx':
        for file in Files:
            df = pd.read_excel(file, index_col=0)
            dict_of_files[str(file)[5:-5]] = df
    return dict_of_files

def merge_two_dicts(x, y):
    z = x.copy()
    z.update(y)
    return z
```

```python
csv_files = read_file_to_dict()
xlsx_files = read_file_to_dict(file_type='xlsx')
data = merge_two_dicts(csv_files, xlsx_files)
```
